##  Analysis of numerical variables


```python
# Create function for Bivariate analysis with Defaulter and Non-Defaulter

def plotbivar(var1,var2):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.scatterplot(x=var1, y=var2,data=ap_data,ax=ax1)
    ax1.set_xlabel(var1)    
    ax1.set_ylabel(var2)
    ax1.set_title(f'{var1} vs {var2} for Non-Defaulters',fontsize=15)
    
    sns.scatterplot(x=var1, y=var2,data=ap_data,ax=ax2)
    ax2.set_xlabel(var1)    
    ax2.set_ylabel(var2)
    ax2.set_title(f'{var1} vs {var2} for Defaulters',fontsize=15)
            
    plt.show()
```

```python
# Bivariate Analysis with AMT_CREDIT & CNT_FAM_MEMBERS

plotbivar('AMT_CREDIT','CNT_FAM_MEMBERS')
```

We can see that the density is in the lower left corner, similar in both the case.
So the people are equally likely to default if the family is small and the AMT_CREDIT is low. We can observe that larger families and people with larger AMT_CREDIT default less often.

```python
# Bivariate Analysis with AMT_GOODS_PRICE & AMT_CREDIT 

plotbivar('AMT_GOODS_PRICE','AMT_CREDIT')
```

## Numerical vs Numerical

```python
def bivariate_numerical(column1, column2):
    plt.figure(figsize=[15,5])

    plt.subplot(1,2,1)
    sns.scatterplot(data=target_0, x=column1, y=column2)
    plt.title('%s' %column1 + ' vs '+ '%s' %column2 + ' for Non-Defaulter \n')
    plt.xticks(rotation=90)

    plt.subplot(1,2,2)
    sns.scatterplot(data=target_1, x=column1, y=column2)
    plt.title('%s' %column1 + ' vs '+ '%s' %column2 + ' for Defaulter \n')
    plt.xticks(rotation=90)

    plt.show()
```
